```python
# selecter
for div in dom.select("a span"):
    # print({"tag" : div.name, "id":div["id"], "text":div.text})
    print({"tag" : div.name, "text":div.text})

# <div id="content" class="pack_group"> -> 아님
# <div id="content" class="pack_group">
#   <h1 class="blind">네이버카페 검색 결과</h1>
#   <div id="main_pack" class="main_pack">
for div in dom.select('#content > .main_pack'): # 나는 id = content, 자식 중에 class = main_pack
    print({"tag" : div.name, "text":div.text})
```

```python
datas = dom.select('#content > div')
for n in range(len(datas)):
    div = dom.select_one('#content > div:nth-of-type({})'.format((n+1)))
    print({'tag' : div.name, "class":div['class'], 'text':div.text})
    print({'tag' : div.name, "class":datas[n]['class'], 'text':div.text})
```

```python
for n in range(1, len(datas)):
    div = dom.select_one('#content > div:nth-of-type({})'.format((n)))
    pre = div.find_previous_sibling()
    
    print({'tag' : div.name, "class":div['class'], 'text':div.text})
    print({'tag' : pre.name, "class":pre['class'], 'text':pre.text}, end='\n\n')
```

```python
for n in range(1, len(datas)):
    div = dom.select_one('#content > div:nth-of-type({}) ~ div.main_pack'.format((n)))
    if (div != None) and (len(div) > 0):
        pre = div[0].find_previous_sibling()
        print(len(div), {'tag':pre.name, 'class':pre['class'], 'text':pre.text})
```
