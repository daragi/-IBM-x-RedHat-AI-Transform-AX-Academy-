```python
from urllib.parse import urlparse, urljoin
from requests import get
from bs4 import BeautifulSoup

seed = 'https://search.naver.com/search.naver?where=nexearch&sm=top_hty&fbm=0&ie=utf8&query=BFS&ackey=gvbz69qv'
base = urlparse(seed)[1]
tovisit = [seed]
visited = []
urls = []

while tovisit:
    url = tovisit.pop(0)
    resp = get(url)
    dom = BeautifulSoup(resp.content, 'lxml')
    visited.append(url)

    for tag in dom.select('a[href]'):
        href = tag.get('href')
        if not href:  # None 또는 빈 문자열일 경우
            continue
        if href.startswith('#') or urlparse(urljoin(seed, href))[1] != base:
            continue

        new_url = urljoin(seed, href)
        if new_url not in visited and new_url not in tovisit and new_url != seed + '/':
            tovisit.append(new_url)
            urls.append(href)
visited
```
