```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
X = np.arange(1, 10, 0.2)
tf1 = np.ones(X.shape[0])
tf2 = X
tf3 = X/X.sum()
tf4 = np.log(1+X)
tf5 = 0.5+(1.0 - 0.5)*X/X.max()
tf6 = 0.0+(1.0 - 0.0)*X/X.max()
plt.plot(X, tf1, label='TF1')
plt.plot(X, tf2, label='TF2')
plt.plot(X, tf3, label='TF3')
plt.plot(X, tf4, label='TF4')
plt.plot(X, tf5, label='TF5')
plt.plot(X, tf6, label='TF6')
plt.axhline()
plt.ylim(-0.1, 2)
plt.legend(loc='upper right')
plt.show()
```

```python
X = np.arange(1, 100, 1)
tf1 = np.ones(X.shape[0])
idf1 = np.ones(X.shape[0])
idf2 = np.log(X.max()/X)
idf3 = np.log(1+X.max()/X)
idf4 = np.log(X.max()/(1+X))
idf5 = np.log((X.max()-X)/X)
plt.plot(X, idf1, label='IDF1')
plt.plot(X, idf2, label='IDF2')
plt.plot(X, idf3, label='IDF3')
plt.plot(X, idf4, label='IDF4')
plt.plot(X, idf5, label='IDF5')
plt.axhline()
plt.legend(loc='upper right')
plt.show()
```

```python
from konlpy.corpus import kobill
Collection = list()
for file in kobill.fileids():
    docid = len(Collection)
    Collection.append((docid, file, kobill.open(file).read()))
Collection
```

```python
from konlpy.tag import Kkma
DTM =list()
for document in Collection:
    termfreq = dict()
    for term in [_ for _ in Kkma().morphs(document[2]) if len(_) > 1]:
        if term not in termfreq:
            termfreq[term] = 1
        else:
            termfreq[term] += 1
    DTM.append(termfreq)
DTM
```

```python
TDM = dict()
for document in Collection:
    docid  = document[0]
    for term, freq in DTM[docid].items():
        if term not in TDM:
            TDM[term] = list()
        TDM[term].append((docid, freq))
TDM
```

```python
from math import log
Weight = dict()
N = len(Collection)
for term, docinfo in TDM.items():
    IDF = log(N/len(docinfo))
    for _ in docinfo:
        docid = _[0]
        maxfreq = max(DTM[docid].values())
        TF = _[1]/maxfreq
        if term not in Weight:
            Weight[term] = list()
        Weight[term].append((docid, TF*IDF))
Weight
```
