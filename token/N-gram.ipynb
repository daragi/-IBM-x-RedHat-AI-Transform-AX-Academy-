```python
import nltk
from nltk.tokenize import sent_tokenize, PunktTokenizer
from nltk.tokenize import sent_tokenize, PunktSentenceTokenizer
from nltk.tokenize import word_tokenize, TreebankWordTokenizer
from nltk.tokenize import word_tokenize, WordPunctTokenizer
from nltk.tokenize import regexp_tokenize, WhitespaceTokenizer, WordPunctTokenizer
from konlpy.corpus import kolaw
from konlpy.tag import Kkma, Komoran, Hannanum, Okt
from nltk.tag import pos_tag
from nltk.probability import FreqDist
```

```python
def ngarm_by_word(t, n=2):
    tokens = list()
    for i in range(len(t) - (n-1)):
        tokens.append(tuple(t[i:i+n]))
    return tokens
tokens = word_tokenize('대한민국의 주권은 국민에게 있다.')
ngarm_by_word(tokens, 1), \
ngarm_by_word(tokens), \
ngarm_by_word(tokens, 3)
```

```python
import re
def ngarm_by_syllable(s, n=2):
    s = re.sub('\n', ' ', s)
    s = re.sub('\s{2,}', ' ',s)
    s = list(re.sub('', '_', s))
    tokens = list()
    for i in range(len(s) - (n-1)):
        tokens.append(tuple(s[i:i+n]))
    return tokens

sentence = '대한민국의 주권은 국민에게 있다.'
ngarm_by_syllable(sentence, 1), \
ngarm_by_syllable(sentence), \
ngarm_by_syllable(sentence, 3), \
ngarm_by_syllable(sentence, 4)
```

```python
corpus = kolaw.open(kolaw.fileids()[0]).read()
tokens = word_tokenize(corpus)
unigram = FreqDist(ngarm_by_word(tokens, 1))
bigram = FreqDist(ngarm_by_word(tokens, 2))
trigram = FreqDist(ngarm_by_word(tokens, 3))

list(zip(unigram.most_common(10), bigram.most_common(10),
         trigram.most_common(10)))
```
