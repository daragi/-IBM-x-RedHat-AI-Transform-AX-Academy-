```python
# 최대공약수, 최대공배수

# 유클리드 호제법
def gcd(a, b):
    
    while b != 0:
        a, b = b, a % b
    return a

def lcm(a, b):
    return (a * b) // gcd(a, b)

# 테스트
a = 48
b = 18

print(f"{a}와 {b}의 최대공약수(GCD):", gcd(a, b))
print(f"{a}와 {b}의 최소공배수(LCM):", lcm(a, b))
```

```python
# 최대 공약수, 최소공배수
# math 라이브러리 이용
import math
num1 = int(input("첫 번째 수를 입력하세요."))
num2 = int(input("두 번째 수를 입력하세요."))

#최대공약수
def solution(num1,num2):
    gcd = math.gcd(num1, num2)
    return gcd
result = solution(num1, num2)
print("두 수의 최대공약수 : {}".format(result))

#최소공배수
def solution2(num1, num2):
    lcm = math.lcm(num1, num2)
    return lcm
result = solution2(num1, num2)
print("두 수의 최소공배수 : {}".format(result))
```
