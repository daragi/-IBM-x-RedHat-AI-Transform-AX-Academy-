```python
import pandas as pd
data = pd.read_csv("./FremontBridge.csv", 
                   index_col = 'Date', 
                   parse_dates=True)
data.head(3)
```

```python
data.head()
```

```python
# 열이름을 단축하고 total 추가
data.columns = ['East', 'West']
data['Total'] = data.eval('West+East')
data.head(5)
```

```python
#요약 통계
data.describe()
```

```python
%matplotlib inline
import seaborn
seaborn.set()
```

```python
data.plot()
plt.ylabel('Hourly Bicycle Count')
```

```python
# 25,000개의 시간별 표본 이해하기 어려움
# 데이터를 주 단위로 리 샘플링
weekly = data.resample('W').sum() #week 단위의 합계로 리샘플링
weekly.plot(style=[':','--','-']) # 첫 번째 열: 점선, 두 번째 열: 대시, 세 번째 열:실선선
plt.ylabel('Weekly bicycle count')
# 4~6월이 교통량 피크임
# 가우시안 그림에서는 다르게 보일 수 있음음
```

```python
# 30일 이동 평균(rolling)
daily = data.resample('D').sum() #day의 함계로 리셈플링
daily.rolling(30, center=True).sum().plot(style=[':','--','-'])
# 30일 짜리 이동 평균을 만듦
# center =True 윈도우를 가운데 정렬해서 양쪽으로 15일씩 계산 -> 중앙정렬값
# 30일치 합계를 그림
plt.ylabel('mean hourly count')
```

```python
# 가우스평활(Gaussian smoothing) 적용
# 가우스 윈도우(Gaussian window) 같은 윈도우 함수를 사용해 롤링 평균을 부드럽게 표현
# 윈도우 폭(50일)과 윈도우내 가우스 폭(10) 지정
daily.rolling(50, center=True,
 win_type='gaussian').sum(std=10).plot(style=[':','--','-'])
# 50일 짜리 슬라이딩 윈도우
# win_type ='gaussian' 가우시안 정규분포 곡선 커널을 사용
# 가까운 날짜일수록 더큰 가중치, 멀수록 적은 영향
# 가중합 계산, std= 10 가우시안 분포의 표준편차
# 가우싀 평활은 정교한 데이터 분석을 위한 , 노이즈 제거 및 outlier에 덜 민감
# 연속성을 유지하여 부드러운 곡선을 유지해줌
```

```python
# 하루의 시간대를 기준으로 한 함수로 평균 통행량을 보고 싶을때
by_time = data.groupby(data.index.time).mean()
# 같은 시간 끼리 그룹핑, 시간대별로 평균값 계산
hourly_ticks = 4*60*60*np.arange(6) # 
# 0 , 14400, 28800 .... => 0시, 4시, 8시, 12시, 16시, 20시
by_time.plot(xticks=hourly_ticks, style=[':','--','-'])
# 출퇴근 시간에 급증하는 교통량
# 출퇴근 시간이 9 to 16임을 알 수 있음
# 출근과 퇴근의 교통량이 다름 
# => 출근의 이동량은 시간대를 폭넓게 잡을 수 있으나, 퇴근은 칼퇴근이 목적임
# 출근 때는 서쪽에서 들어오고 퇴근은 동쪽으로 나감 
# => 교대 근무 혹은 지역별 주거 밀집 지역에 따른 교통량이 생성됨
# ex) 서울(출근 지역)과 경기도 광주(주거 지역)
```

```python
# 요일별 통행량은?
by_weekday = data.groupby(data.index.dayofweek).mean()
by_weekday.index = ['Mon','Tues','Wed','Thurs','Fri','Sat','Sun']
by_weekday.plot(style=[':','--','-'])
# 사람들의 생활패턴에 따라 달라짐
# 월, 금 휴가 => 불목(?) => 주말에는 타지역의 본가, 평일에는 직장 근처에서 거주
# 토요일>일요일 => 토요일 출근 및 운동 레저 활동
```

```python
# 주중과 주말의 시간대별 추이
# 데이터를 주말을 표시하는 플래그와 시간대별로 분류
weekend = np.where(data.index.weekday <5, 'Weekday', 'Weekend')
by_time = data.groupby([weekend, data.index.time]).mean()
```

```python
# 다중 서브플롯
import matplotlib.pyplot as plt
fgs, ax = plt.subplots(1, 2, figsize=(12,7))
by_time.loc['Weekday'].plot(ax=ax[0], title='Weekdays',
 xticks=hourly_ticks, style=[':','--','-'])
by_time.loc['Weekend'].plot(ax=ax[1], title='Weekends',
 xticks=hourly_ticks, style=[':','--','-'])
# 앞에서의 그래프보다 분봉이 홀쭉해짐 => 주말의 교통량이 빠졌기 때문에
```

```python
# ( 시애틀의 자전거 통행량 예측 )
# 미 기상청의 데이터세트와 연결
# 선형 회귀를 사용해 날씨와 다른 정보를 자전거 수와 관련 짓고
# 이 가운데 한 가지 모수의 변화가 해당 일자에 자전거를 탄 사람의 수에 어떻게 영향을 주는 지 추정
import pandas as pd
counts = pd.read_csv('FremontBridge.csv', index_col='Date', parse_dates=True)
# parse_dates = True => 지정된 열을 datetime형식으로 파싱
weather = pd.read_csv('BicycleWeather.csv', index_col='DATE', parse_dates=True)
```

```python
# 일별 총 자전거 통행량을 계산해서 별도의 DataFrame에 넣음
daily = counts.resample('d').sum()
daily['Total'] = daily.sum(axis=1)
daily = daily[['Total']] # 다른 열 삭제
daily
```

```python
daily.head(3)
```

```python
# 요일을 나타내는 열 추가
days = ['Mon', 'Tue', 'Wed', 'Thu', 'Fri', 'Sat', 'Sun']
for i in range(7):
 daily[days[i]] = (daily.index.dayofweek == i).astype(float)
daily.head(3)
```

```python
# 공휴일 추가
from pandas.tseries.holiday import USFederalHolidayCalendar
cal = USFederalHolidayCalendar()
holidays = cal.holidays('2012', '2016')
daily = daily.join(pd.Series(1, index=holidays, name="holiday"))
# print(daily.index[""]) #2012년도의 휴일 날짜 샘플 출력하기 -과제
print(holidays)
```

```python
# 일조시간에 자전거를 타는 사람
def hours_of_daylight(date, axis=23.44, latitude=47.61): #axis=경도 ,latitude=위도
 # 해당 날짜의 일조시간 계산
 days = (date - datetime.datetime(2000, 12, 21)).days
 m = (1. -np.tan(np.radians(latitude))
 * np.tan(np.radians(axis) * np.cos(days * 2 * np.pi / 365.25)))
 return 24 * np.degrees(np.arccos(1 - np.clip(m, 0, 2))) / 180.
daily['daylight_hrs'] = list(map(hours_of_daylight, daily.index))
daily[['daylight_hrs']].plot()
plt.ylim(8, 17)
```

```python
# 데이터에 평균 기온과 전체 강수량 추가
# 인치 단위의 강수량과 더불어 날이 건조했는지(강수량이 0) 알려주는 플래그 추가
# 기온은 섭씨 1/10도 단위, 섭씨 1도 단위로 변환
weather['TMIN'] /= 10
weather['TMAX'] /= 10
weather['Temp (c)'] = 0.5 * (weather['TMIN'] + weather['TMAX'])
# 강수량은 1/10mm 단위; 인치 단위로 변환
weather['PRCP'] /= 254
weather['dry day'] = (weather['PRCP'] == 0).astype(int)
daily = daily.join(weather[['PRCP', 'Temp (c)', 'dry day']])
```

```python
daily.head(4)
```

```python
# 첫날 부터 증가하는 계수기를 추가해 몇 해가 지났는지를 측정
# 관측된 일별 통행량이 연도별로 증가하거나 감소하는지를 측정
daily['annual'] = (daily.index - daily.index[0]).days / 365.
daily.head()
```

```python
daily.head()
```

```python
print(daily.loc['2013-10-03'])
```

```python
# 널값 행은 제거
daily.dropna(axis=0, how='any', inplace=True)
column_names = ['Mon', 'Tue', 'Wed', 'Thu', 'Fri', 'Sat', 'Sun',
 'holiday', 'daylight_hrs', 'PRCP', 'dry day','Temp (c)', 'annual']
X = daily[column_names] # 문제지
y = daily['Total'] # 답안지
X.head(3)
```

```python
#인공지능(머신러닝 : 선형회귀 알고리즘 사용)
from sklearn.linear_model import LinearRegression
model = LinearRegression(fit_intercept=False)
```

```python
# 학습시키기
model.fit(X, y)
```

```python
# 인공지능 프로그램 검증 : 결과예측하기
daily['predicted'] = model.predict(X)
daily.head()
```

```python
#총 자전거 통행량과 예상 자전거 통행량 비교
daily[['Total', 'predicted']].plot(alpha=0.5)
```

```python
# 자전거 통행량에 각 특징이 얼마나 기여하는지 추정하는 선형 모델 계수
params = pd.Series(model.coef_, index=X.columns) #coef : 가중치, 계수
params
```

```python
# 불확실성에 대한 척도 없이는 해석이 어려움
# 데이터의 부트스트랩 표본 재추출(bootstrap resampling)을 사용하여 불확실성을 계산
from sklearn.utils import resample
np.random.seed(1)
err = np.std([model.fit(*resample(X, y)).coef_
 for i in range(1000)], 0 )
```

```python
# 추정된 오차를 가지고 결과를
print(pd.DataFrame({'effect' : params.round(0), 'error' : err.round(0)}))
```
