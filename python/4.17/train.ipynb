```python
sampleDf = df.iloc[0:5, 1:3]
sampleDf = df[["Survived", "Pclass", "Age", "SibSp", "Parch", "Fare"]]
# 분석할 열 추출
print(sampleDf)
print(sampleDf.corr())
print(sampleDf.groupby("Survived").mean())
```

```python
mask = (df.Sex == "female")
print(mask) # 그냥 시리즈 출력
print([[mask]]) # 이중 리스트로 감싼 구조
```

```python
sampleDf["New"] = 0
sampleDf["Family"] = sampleDf["SibSp"] + sampleDf["Parch"] 
# 가족 수 계산 family 라는 열 생성
print(sampleDf)
sampleDf = sampleDf.drop(labels="New", axis=1) #drop() = new라는 열 삭제, 열기준
print(sampleDf)
sampleDf.drop(columns=["Family"], inplace=True) # family 열 삭제, inplace=true 원본에서 삭제제
print(sampleDf)
```

```python
data = np.arange(len(sampleDf.columns)).reshape(1, -1) #sampleDf의 열 개수 = 6개
                                                       #reshape을 통해 1행 N열 행태로 바꿈꿈
print(data)
number = pd.DataFrame(data, columns=sampleDf.columns)  #data = 012345 를 열에 넣음
print(number)
rdf = pd.concat([sampleDf, number], axis=0) # sampleDf 아래에 넘버를 행기준으로 붙임
print(rdf)
rdf = rdf.reset_index(drop=True) #중복되는 인덱스를 새 인덱스로 재설정 -> 중복 제거거
print(rdf)
```

```python
df.rename({"Age" : "나이"}, axis=1) # age 컬럼 명을 나이로 변환, 원본 반영x , inplace=True 없음
age_mean = sampleDf.Age.mean() # age의 평균값을 age_mean 변수로 저장
print(age_mean)
sampleDf["Age"] = sampleDf["Age"].fillna(value=age_mean) # fillna를 통해 NaN 값을 age_mean으로 채움
print(sampleDf[888:]) #888번 인덱스부터 출력
```

```python
def f(x) : 
    return len(str(x))
print(df.applymap(f)) # applymap은 dataframe의 모든 원소에 함수 f 적용
print(f(df)) 
```

```python
print("########## map() test ###########") #map은 series의 전용으로 쓰임
print(df.Sex)
print(df.Sex.map(lambda x : 0 if x == "female" else 1))
# 여자는 0, 남자는 1값으로 리턴 하는 map을 사용 => lambda를 사용해 모든 값에 함 수적용용
print(df.Sex.map({"female":10, "male" : 20})) 
# map은 딕셔너리를 통해 지정된 값(key)에 대한 value를 반환
```

```python
def f2(x, n) :  
    return x ** n
print(df[["Age", "SibSp", "Fare"]].apply(f2, args=[2]))
# 세 숫자형 열에 apply를 적용해 , 모든 값에 f2 함수 적용 = 제곱
# args=[2]: n=2를 넘겨 f2 함수를 호출
df.to_csv("./titanic.csv")
df.to_clipboard() #df를 클립보드에 복사

for dformat in dir(df): #모든 속성과 메서드 리스트
    if dformat.startswith("to_"): # to(저장 또는 변환 관련 함수)를 필터링 
        print(dformat)
```
