```python
data = sns.load_dataset("iris")
sns.jointplot(x="sepal_length", y="sepal_width", data=data)
plt.title("JointPlot by Iris Sepal width / length")
plt.show()
```

```python
sns.jointplot(x="sepal_length", y="sepal_width",
              data=data, kind="kde", color='blue', space=0)
plt.title("   joint", y=1.02)
plt.show()
```

```python
sns.jointplot(x="sepal_length", y="sepal_width",
              data=data, kind="hex", color='green')
plt.title('joint sepal_length with sepal_width')
plt.show()
```

```python
sns.pairplot(data, hue="species", markers=['o','s','D'])
plt.title("Iris Pair Plot, Hue   ")
plt.show()

# sepal_width     
# 3,4,12,15 +   
#       ( )
```

```python

titanic =sns.load_dataset('titanic')
titanic_size= titanic.pivot_table(index= "age",
                                  columns='survived',
                                  aggfunc='size') # 
titanic_size
```

```python
plt.figure(figsize=(5,15))

sns.heatmap(titanic_size, cmap=sns.light_palette("red", as_cmap=True), 
            annot=True, fmt='.2f')

plt.title("Heatmap")
plt.show()
#   ,   1:1.7 
#          
#  ,    (=third)
# 2 , , , , 
# 1 ,  
```

```python
tips =sns.load_dataset('tips')
sns.violinplot(x='day', y='total_bill', data=tips)
plt.title("total tips / group by days to violin plot")
plt.show()
```

```python
np.random.seed(0)
sns.violinplot(x='day',y='total_bill',data=tips)
sns.stripplot(x="day",y="total_bill", data=tips, jitter=True) #
sns.swarmplot(x='day',y='total_bill',data=tips) #  
plt.title("Total bills of weeks")
plt.show()
```

```python
sns.boxplot(x='day', y='total_bill', hue='sex', data=tips)
plt.title("box plot")
plt.show()
```

```python
flights = sns.load_dataset("flights")
```

```python
plt.figure(figsize=(12,6))
flights_passengers = flights.pivot(index="month",
                                   columns="year",
                                   values="passengers")

sns.heatmap(
    flights_passengers, cmap=sns.light_palette("red", as_cmap=True), annot= True, fmt='d', linewidths=1)
plt.title("Total Year/Month")
```

```python
def sinplot(flip=1):
    x = np.linspace(0,14, 100)
    for i in range(1,7):
        plt.plot(x, np.sin(x+i*.5)*(7-i)*flip)
sns.set_style("dark")
plt.title("Sin plot")
sinplot()
```
