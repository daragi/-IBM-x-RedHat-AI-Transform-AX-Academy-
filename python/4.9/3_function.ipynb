```python
def fib(n):
    """Print a Fibonacci series up to n."""
    a,b = 0,1
    while a <n:
        print(a,end=' ')
        a,b = b,a +b
fib(2000)
```

```python
print(fib(0))
```

```python
def fib2(n):
    """Print a Fibonacci series up to n."""
    a,b = 0,1
    result = []
    while a <n:
        print(a,end=' ')
        result.append(a)
        a,b = b,a +b
    return result
```

```python
print(fib2(30))
```

```python
f100 = fib2(100)
print("result :",f100)
```

```python
def ask_ok(prompt,retries=4,reminder='Please try again!'):
    while True:
        ok =input(prompt)
        if ok in ('y','ye','yes'):
            return True
        if ok in ('n','no','nop','nope'):
            return False
        retries = retries -1
        if retries <0:
            raise ValueError('invalid user response')
        print(reminder)
#
```

```python
ask_ok('yes')
```

```python
ask_ok('입력:', 3, '다시 입력:')
```

```python
def f(a,L=[]):
    L.append(a)
    return L
print(f(1))
print(f(2))
print(f(3))
# 파라미터 (함수의 매개변수)에 넣으면 변환이 가능
```

```python
def f(a,L=[]):
    L.append(a)
    return L
print(f(1))
print(f(2, [1,2])) 
# [1,2] o, [1,2,2]?? why? 
# [1,2]는 파라미터로 할당이 불가능하기에 외부에 들어온 전역변수로 인식
print(f(3)) # [1,2,2,3] o, [1,3]?? why?
# f(1)은 지역변수만 있기 때문에 append로 리스트가 쌓이는데
# f(2, [1,2])는 전역변수가 섞여서 새로운 리스트가 만들어짐 그 과정에서 함수의 몸만 빌림
# f(3)은 지역변수만 사용하기에 원래의 1, 3이 쌓임 => [1,3]
```

```python
def f(a,L=None):
    if L is None:
        L = []
    L.append(a)
    return L
print(f(1))
print(f(2))
print(f(3))
```

```python
def parrot(voltage,state='a stiff',action='voom',type='Norwegian Blue'):
    print("-- This parrot wouldn't",action,end=' ')
    print("if you put",voltage,"volts through it.")
    print("-- Lovely plumage, the",type)
    print("-- It's",state,"!")
```

```python
parrot(1000)
parrot(voltage=1000) 
parrot(voltage=1000000,action='VOOOOOM') 
parrot(action='VOOOOOM',voltage=1000000) 
parrot('a million','bereft of life','jump') 
parrot('a thousand',state='pushing up the daisies')
# parrot()  voltage 값은 default값이 없기에 반드시 입력해주어야 함, 빈 값은 불가
# parrot(voltage=5.0, 'dead') keyword 인자 뒤에 , 위치 인자가 오는 것은 불가능
# parrot(110, voltage=220) 함수 하나에 voltage 값이 두 개 들어감
# parrot(actor='John Cleese') 매개 변수가 엉뚱한 것이 들어감
```
