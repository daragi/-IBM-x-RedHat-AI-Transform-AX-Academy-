```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sn
```

```python
x = np.array([1,2,3,4])
y = np.array([2,4,6,8])

plt.plot(x)
plt.plot(y)
plt.show()
```

```python
plt.plot(x)
plt.plot(x, x**2)
plt.show()
```

```python
x = np.arange(0, 5, 0.2)
plt.plot(x, x**2, "ro") # red, o= dot
plt.plot(x, x**3, "b^") # blue, ^ = triangle
plt.plot(x, x**4, "gs") # green, s = square
plt.show()
```

```python
x = np.array([1,2,3,4])
y = np.power(x,2)
plt.bar(x,y)
plt.show()
plt.barh(x,y) #가로형 막대 그래프 = barh
plt.show()
```

```python
np.random.seed(0) # rand의 결과를 seed를 통해 랜덤 값을 고정
x = np.random.rand(50)
y = (x * 10) + np.random.rand(50)
plt.scatter(x,y) # dot은 생략돼있음 => plt.plot(x,y,"o")
plt.show()
```

```python
np.random.seed(0)
spread = np.random.rand(50) * 100
center = np.ones(25) * 50
flier_high = np.random.rand(10) * 100 + 100
flier_low = np.random.rand(10) * -100
data = np.concatenate((spread, center, flier_high, flier_low))
plt.boxplot(data)
plt.show()
```

```python
data = np.array([1,2,3,4])
labels = ["A","B","C","D"]
plt.pie(data, labels=labels)
plt.show()
```

```python
np.random.seed(0)
mu, sigma = 100,15
data = mu + sigma * np.random.randn(10000)
plt.title("title")
plt.xlabel("xlabel")
plt.ylabel("ylabel")
plt.xticks([0,50,100,150]) #x축의 범위 설정
plt.yticks([0,1200,2500]) #y축의 범위 설정
plt.hist(data)
plt.grid() #격자 무늬
plt.show()
```

```python
x1 = np.arange(5)
x2 = np.power(x1,2)
x3 = np.power(x1,3)
plt.plot(x1,"ro" ,x2,"bs",x3,"g^")
plt.title("title")
plt.xlabel("xlabel")
plt.ylabel("ylabel")
plt.xticks([0,2,4]) #x축의 범위 설정
plt.yticks([0,30,60]) #y축의 범위 설정
plt.legend(["x1","x2","x3"]) #범례
for feature in [x1, x2, x3]:
    for i, data in enumerate(feature):
        plt.text(i, data+1.5, data)
plt.show()
```
