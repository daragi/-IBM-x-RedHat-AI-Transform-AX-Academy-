```python
# 스칼라 = 0차원
# 스칼라의 집합 = series(1차원)
# 시리즈의 집합 = DataFrame(2차원)
# 시리즈는 라벨이 붙었음
# 시리즈 안에는 한가지의 자료형으로 통일되어야 한다(숫자는 숫자끼리, 문자는 문자끼리)
```

```python
data = [1,2,3,4]
index = ["A","B","C","D"]
series1 = pd.Series(data=data, index=index) 
series1
```

```python
data = {"col1":[1,2,3,4], "col2":["a","b","c","d"]}
index = ["A","B","C","D"]
df1 = pd.DataFrame(data=data, index=index) #pd.DataFrame(data내용 + col, index)
df1
```

```python
data = [[1,"a"],
        [2,"b"],
        [3,"c"],
        [4,"d"]]
columns = ["col1", "col2"]
df2 = pd.DataFrame(data=data, columns=columns) #index 생략 시 자동 생성
print(df2)
df3 = pd.DataFrame(data, columns=columns, index=index)
print(df3)
```
