```python
import numpy as np
import time
```

```python
# 데이터의 개수
n = 1000
# 피처의 개수
m = 10000
```

```python
# x 행렬의 크기는 1000 x 10000
X = np.random.rand(n,m)
# W 행렬의 크기는 1000 x 1
W = np.random.rand(n,1)
X,W
```

```python
# 반복문 사용한 코드
t1 = time.time()
Z1 = np.zeros((1,m)) #np.zeros(1행, m개의 열)
for i in range(X.shape[1]):
    for j in range(X.shape[0]):
        Z1[0][i] += W[j]*X[j][i]
print("반복문 사용한 코드의 속도 : ",(time.time()-t1)*1000,"ms")
```
