```python
# 리스트는 스택 방식 - last in, first out 구조
# 그렇다면 리스트 인덱싱이나, pop의 인덱싱 기능은 어떻게 설명할 것인가?
# append = > stack
# pop() => stack
# pop(i) => queue
stack = [3,4,5]
stack.append(6)
stack.append(7)
stack
stack.pop()
stack
```

```python
# 리스트를 queue로도 사용 가능
from collections import deque 
# deque 앞과 뒤에서 데이터를 처리할 수 있는 양방향 자료형으로 queue 처럼 써도 되고
# stack 처럼 써도 된다.
queue = deque(["Eric", "John","Michael"])
queue.append("Terry")
queue.append("Graham")
queue.popleft()
print(queue.popleft())
queue
```
