```python
def standard_arg(arg): # 일반매개변수
    print(arg)
    
def pos_only_arg(arg,/):  # arg, / 는 앞의 매개 변수를 위치로 받는다는 뜻
    print(arg)

    # *이면 뒤가 list,set을 받는 키워드매개변수고 **이면 뒤가 dict를 받는 키워드매개변수
    # ** 뒤에 변수는 dictionary 를 언패킹 한다는 뜻도 있음 
def kwd_only_arg(*,arg): # 키워드 매개변수
    print(arg)

standard_arg("standard_arg(arg)") # 자유로운 일반매개변수~
standard_arg(arg = "standard_arg(arg)") #키워드 매개변수


pos_only_arg("pos_only(arg,/):") #argument 가 위치 매개변수가 됨

# 키워드 매개변수, 값을 할당하므로 변하지 않됨, arg는 키워드매개 변수가 됨
kwd_only_arg(arg ="kwd_only(*,arg):")

# 공식 순서
# 기본값 인자 -> 가변매개변수(*args) -> 키워드 매개변수 -> 가변 키워드매개변수(**kwargs) , 위치 인자의 순서는 가변매개변수 앞으로
```

```python
# /,name을 위치 인자로 변화, 가변 키워드 매개변수 = > name에 중복으로 할당되지 않음
def foo(name,/,**kwds): 
    return 'name' in kwds
foo(1, **{'name':2})
```

```python
def write_multiple_items(file,separator,*args): # 가변 매개변수
    # file.write(separator.join(args))
    print(separator.join(args))
```

```python
def concat(*args, sep="/"): # 가변매개변수, sep을 통해 변환
    return sep.join(args)

concat("earth","mars","venus")
concat("earth","mars","venus", sep=".")
```
