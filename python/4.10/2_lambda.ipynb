```python
pairs =[(1,'one'),(2,'two'),(3,'three'),(4,'four')]

pairs.sort(key=lambda pair:pair[1]) 
# 인덱싱 1이 알파벳으로 들어감으로 알파벳 순서대로 정렬
pairs
```

```python
pairs.sort(key=lambda pair:pair[0]) # 숫자대로 정렬
pairs
```

```python
pairs_ =[(1,'1'),(2,'12'),(3,'3'),(4,'21')] # 앞자리의 1,2,3 순서를 쫓음
pairs_.sort(key=lambda pair:pair[1])
pairs_ 
# pairss. => 객체
# 멤버변수 => 리스트 내의 key 가 되는 값
# 메소드 => .sort ex) .append .pop .insert .remove .clear
```
