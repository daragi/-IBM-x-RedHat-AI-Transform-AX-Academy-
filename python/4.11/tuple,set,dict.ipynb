```python
# zip(arg) = 패킹 => zip(*arg) = 언패킹
# 패킹은 언패킹의 멱멱수 이므로 멱함수 알고리즘이 존재해야함
questions = ['name','quest','favorite color']
answers = ['lancelot','the holy grail','blue']
for q,a in zip(questions, answers):
    print('What is your {0}? It is {1}.'.format(q,a))
```

```python
import math

raw_data = [56.2,float('NaN'),51.7,55.3,52.5,float('NaN'),47.8] 
#NaN = not a number, type은 float
filtered_data = []
for value in raw_data:
    if not math.isnan(value): # isnan수학함수는 NaN 값을 체크, 출력하는 메소드
        filtered_data.append(value)
filtered_data

# nan 필요없는 값임에도 불구하고 삭제시키지 않는 이유 
# raw data(원형 데이터)를 손상시킬 경우
# 다음 시행 때 삭제한 것이 영향을 미침 , 원형 데이터 보존이 중요
# 심화
# NaN 포함 값을 연산하거나 데이터 분석을 할 때 none 값을 outlier(이상치)로 규정
# 회귀분석에서는 원 데이터를 따로 분석하고 , outlier에 대해선 잔차 분석을 진행
```

```python
import math

raw_data = [56.2,float('NaN'),51.7,55.3,52.5,float('NaN'),47.8] 
filtered_data = []
for value in raw_data:
    if math.isnan(value): # not isnan의 반대 버전
        filtered_data.append(value)
filtered_data
```

```python
# A and not B or C 결과값
# T         T    T  => T   and     (not T) = F => T and F => F  => F or T = T
#                F F  
#           F      T
# A and (not B) or C  => not 은 가장 높은 연산 순위를 가지므로 괄호를 씌움
# (A and (not B)) or C
```

```python
string1,string2,string3 = '','Trondheim','Hammer Dance' 
# 공백은 false로 생각, value 값이 있으면 true로 생각
non_null = string1 or string2 or string3 # or 앞에 있는 값의 연산 순위가 더높음
non_null
```

```python
string1,string2,string3 = '','Trondheim','Hammer Dance'
non_null = string1 and not string2 or string3
non_null
```
