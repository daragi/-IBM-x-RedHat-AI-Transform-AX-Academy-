```python
def fib2(n):
    result = []
    a,b = 0,1
    while a<n:
        result.append(a)
        a,b = b,a + b
    return result
resultList = fib2(1000)
# fibonacci.py 파일을 가져와 fibonacci 모듈을 fib3로 명명
import fibonacci as fib3
print(fib3.__name__) # 점프투파이썬 p.217 참고 ㅎㅎ
resultList[:] = []
resultList = fib3.fib(1000)
print(resultList)
# import -> '동적' 라이브러리 호출(dll)
# include -> (C,C++) 언어 정적 라이브러리 호출(sll)
```
